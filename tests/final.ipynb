# Problem 9
Define a function `myStringSum3` that accepts one argument (a string) that contains numbers separated by semicolons (`';'`). The function should return the sum of all the numbers.

````python
def myStringSum3(string):
    nums = string.split(';')
    total = 0
    for num in nums:
        total += int(num)
    return total
````

# Problem 15

Define a function `wordsWithVowels()` that accepts one argument (a string) and returns a list of words that contain a vowel (`['a', 'e', 'i', 'o', 'u']`).

`wordsWithVowels('try to fly by the house')` should return `['to', 'the', 'house']`.

- Note: we define the `hasVowel()` function for your convenience

*(7 points)*

````python
def hasVowel(word):
    for vowel in ['a', 'e', 'i', 'o', 'u']:
        if vowel.upper() in word.upper():
            return True
    return False

# write your code for problem 15 here

def wordsWithVowels(string):
    return [word for word in string.split(' ') if hasVowel(word) == True]
````

Define a function named `shortWordCount` that accepts one argument (a sentence, as a string) and returns the number of words in that string that are shorter than 5 characters. Your code **must** use either list comprehension or the `filter()` function to do this. *(7 points)*

````python
def shortWordCount(sentence):
    if len(sentence) > 0:
        lst = list(filter(lambda word: len(word) < 5, sentence.split(' ')))
        return len(lst)
    else:
        return 0
````

# Problem 20

Define a function `getCharFreq()` that accepts one argument (a string, `sentence`) and returns a dictionary where the keys are characters in `sentence` and the values are how often that character appears.

*(12 points)*

````python
def getCharFreq(sentence):
    d = {}
    for achar in sentence:
        try:
            d[achar] = d[achar] + 1
        except:
            d[achar] = 1
    return d
````

You can also assume that every vowel appears at least once and that there is just one least frequent vowel (no ties).

# Problem 24

The [MetaWeather API](https://www.metaweather.com/api/) allows you to get weather data for a given location. Define a function named `rainToday` that accepts one argument (a city name) and uses the MetaWeather API to determine if it is raining in that city. We count "raining" as either (1) Thunderstorm, (2) Heavy Rain, (3) Light Rain, or (4) Showers. Your function should return either `True` (it is raining in that city) or `False` (it is not raining in that city).

- Note 1: The MetaWeather API only works for a limited set of cities
- Note 2: You wil need to make two calls to the API; one to search for the city's location ID (woeid) and another to get the weather for that location
- Note 3: Use the first element of the consolidated weather to represent "today"
- Note 4: Use `requests_with_caching`


*(25 points)*

````python
#Defining search function
def rainToday(city_name):
    import requests_with_caching
    rain = ['t', 'hr', 'lr', 's']
    
    # Retrieve ID
    parameters = {"query": city_name}
    response = requests.get("https://www.metaweather.com/api/location/search/", params = parameters)
    results = json.loads(response.text)
    city_id = results[0]['woeid']
    
    # Retrieve weather
    weather = requests.get(f"https://www.metaweather.com/api/location/{city_id}")
    results_w = json.loads(weather.text)
    
    # Rain?
    weather_today = results_w['consolidated_weather'][0]['weather_state_abbr']
    if weather_today in rain:
        return True
    else:
        return False
````

````python
assert rainToday('Detroit') == True
assert rainToday('San Francisco') == False
assert rainToday('London') == True
````
